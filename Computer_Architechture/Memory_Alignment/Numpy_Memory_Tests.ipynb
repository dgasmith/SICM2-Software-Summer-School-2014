```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import timeit
from collections import defaultdict
%pylab inline

#Make images slightly bigger
matplotlib.rcParams['savefig.dpi'] *= 1.5

#Maximum memory in GB also used to modify speed of tests
MAXIMUM_MEMORY = 0.01

#Maximum doubles size
dim_array_max = int(MAXIMUM_MEMORY*1E9/8)
dim_matrix_max = int(dim_array_max**0.5)
```

#Summation of matrices
Summing the first or second dimensions of a matrix should return different timings due to memory alignment:

Sum in contigous memory:
$$d_i = \sum_{j} A_{ij}$$

Sum in non-contigous memory:
$$d_j = \sum_{i} A_{ij}$$

A simply numpy script to produce this is shown below.

```python
sample_size = 50
sample_count = 20

array_sizes = np.linspace(100, dim_array_max, sample_count)
matrix_sizes = np.linspace(100, dim_matrix_max, sample_count)

arr_size = []
f_times = []
c_times = []

for x in matrix_sizes:
    x = int(x)
    arr_size.append(x)

    # Time c-array sum over first axis
    carray = np.random.rand(x, x)
    ctime = timeit.timeit('carray.sum(axis=1)',
                      'import numpy as np; from __main__ import carray',
                      number=sample_size)/(x*1E3)
    c_times.append(ctime)
    
    # Time c-array sum over second axis
    ftime = timeit.timeit('carray.sum(axis=0)',
                      'import numpy as np; from __main__ import carray',
                      number=sample_size)/(x*1E3)
    f_times.append(ftime)
    

# Setup figure
fig, ax = plt.subplots(1, 1)
ax.set_ylabel('1D sum Average Time (ms)')
ax.set_xlabel('Array Size')

# Add plots to figure
ax.plot(np.arange(len(f_times)), f_times, label='Sum over contiguous memory')
ax.plot(np.arange(len(c_times)), c_times, label='Sum over non-contiguous memory')

# Add legend
ax.legend(loc=2)
```

#Timing of different operations
Different operations require different amounts of flops.

```python
sample_size = 50
sample_count = 10

array_sizes = np.linspace(1E4, dim_array_max, sample_count)
matrix_sizes = np.linspace(10, dim_matrix_max, sample_count)

times = defaultdict(list)

def time_arr_operation(operation, number=10):
    return timeit.timeit(operation,
        'import numpy as np; from __main__ import arr',
         number=number)
    

for x in array_sizes:
    x = int(x)
    arr = np.random.rand(x)
    GB_Size = arr.nbytes/1E9
    
    divisor = (arr.nbytes/x)
    times['add'].append(time_arr_operation('arr+arr', number=sample_size)/divisor)
    times['mul'].append(time_arr_operation('arr*arr', number=sample_size)/divisor)
    times['div'].append(time_arr_operation('arr/arr', number=sample_size)/divisor)

        
# Setup figure
fig, ax = plt.subplots(1, 1)
ax.set_ylabel('Average execution time per element')
ax.set_xlabel('Array Size')

for key in times.keys():
    ax.plot(np.arange(sample_count), times[key], label=key)
    
# Add legend
ax.legend(loc=2)
```
